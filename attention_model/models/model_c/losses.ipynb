```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_results(training_loss, validation_loss):
    plt.plot(training_loss, label="Training Loss")
    plt.plot(validation_loss, label="Validation Loss")
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel("MSE")
    plt.show()
```

# Losses

```python
training_loss = np.load('./log/training_losses.npy', allow_pickle=True)
validation_loss = np.load('./log/validation_losses.npy', allow_pickle=True)

plot_results(training_loss,validation_loss)
```
