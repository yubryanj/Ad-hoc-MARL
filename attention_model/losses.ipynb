```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_results(training_loss, validation_loss):
    plt.plot(training_loss, label="Training Loss")
    plt.plot(validation_loss, label="Validation Loss")
    plt.legend()
    plt.xlabel("Epochs")
    plt.ylabel("MSE")
    plt.show()
```

# Model A

```python
training_loss = np.load('./models/model_a/log/training_losses.npy', allow_pickle=True)
validation_loss = np.load('./models/model_a/log/validation_losses.npy', allow_pickle=True)

plot_results(training_loss,validation_loss)
```

## Model B

```python
training_loss = np.load('./models/model_b/log/training_losses.npy', allow_pickle=True)
validation_loss = np.load('./models/model_b/log/validation_losses.npy', allow_pickle=True)

plot_results(training_loss,validation_loss)
```

```python
training_loss = np.load('./models/model_c/log/training_losses.npy', allow_pickle=True)
validation_loss = np.load('./models/model_c/log/validation_losses.npy', allow_pickle=True)

plot_results(training_loss,validation_loss)
```
